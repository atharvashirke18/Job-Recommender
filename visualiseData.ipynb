```
import pandas as pd
import matplotlib.pyplot as plt

BASE_DIR = "/home/lucifer/projects/Job-Recommender/datasets"
```

```
applications = pd.read_csv(f"{BASE_DIR}/applications.csv")
candidates = pd.read_csv(f"{BASE_DIR}/candidates.csv")
jobs = pd.read_csv(f"{BASE_DIR}/jobs.csv")

print("Applications:", applications.shape)#Interactions
print("Candidates:", candidates.shape)#users
print("Jobs:", jobs.shape)#items
```

```
dataset_names = ["Applications", "Candidates", "Jobs"]
dataset_sizes = [
    len(applications),
    len(candidates),
    len(jobs)
]

plt.figure()
plt.bar(dataset_names, dataset_sizes)
plt.title("Dataset Size Comparison")
plt.xlabel("Dataset")
plt.ylabel("Number of Records")
plt.show()
```

```
splits = ["Train", "Cross Validation", "Test"]
percentages = [70, 20, 10]

plt.figure()
plt.bar(splits, percentages)
plt.title("Dataset Split Ratio")
plt.xlabel("Split Type")
plt.ylabel("Percentage")
plt.show()
```

```
# change column name if needed
company_col = "company"

top_companies = jobs[company_col].value_counts().head(10)

plt.figure()
top_companies.plot(kind="bar")
plt.title("Top 10 Companies by Job Postings")
plt.xlabel("Company")
plt.ylabel("Number of Jobs")
plt.show()
```

```
job_application_counts = applications["job_id"].value_counts()

plt.figure()
plt.hist(job_application_counts, bins=50)
plt.title("Applications per Job Distribution")
plt.xlabel("Number of Applications")
plt.ylabel("Number of Jobs")
plt.show()
```

```
candidate_application_counts = applications["candidate_id"].value_counts()

plt.figure()
plt.hist(candidate_application_counts, bins=50)
plt.title("Applications per Candidate Distribution")
plt.xlabel("Number of Applications")
plt.ylabel("Number of Candidates")
plt.show()
```

```
print("Missing values in Applications:")
print(applications.isnull().sum())

print("\nMissing values in Candidates:")
print(candidates.isnull().sum())

print("\nMissing values in Jobs:")
print(jobs.isnull().sum())
```
